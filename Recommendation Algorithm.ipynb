```python
def max_diff(wind_speeds):
    diff = 0
    for i in range(1, len(wind_speeds)):
        if abs(wind_speeds[i] - wind_speeds[i-1]) > diff: 
            diff = abs(wind_speeds[i] - wind_speeds[i-1])
            
    return diff

print(max_diff([2,5,16,3,7]))
```

```python
#[(100, 'E', 6), (300, 'E', 1), (600, 'E', 7)]
def analyse_single_window(list_of_tup):
    wind_speeds = [tup[2] for tup in list_of_tup]
    string_template = ["Ground Wind Speed (kts): ", "Average Wind (Kts): ", "Max Wind (Kts): "]
    string_template[0] += str(wind_speeds[0])
    string_template[1] += str(round(sum(wind_speeds)/len(wind_speeds),3))
    string_template[2] += str(max(wind_speeds))
    
    return "\n".join(string_template)
    
print(analyse_single_window([(100, 'E', 6), (300, 'E', 1), (600, 'E', 7)]))
    
```
